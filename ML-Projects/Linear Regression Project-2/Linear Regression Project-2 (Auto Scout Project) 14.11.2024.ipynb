```python
def outliers(df):
    
    numeric_df = df.select_dtypes(include='number')
    
    for col in df.select_dtypes(include='number').columns:

        Q1 = df[col].quantile(0.25)

        Q3 = df[col].quantile(0.75)

        IQR = Q3 - Q1

        outliers = numeric_df[(numeric_df[col] < (Q1 - 3 * IQR)) | (df[col] > (Q3 + 3 * IQR))]

        

        print(f"{col} - Outliers:")

        print(f"Number of Outliers: {outliers.shape[0]}")

        print(outliers[[col]])  

        print("---------------------------")
```

### Outlier Detection and Analysis

Outliers have been identified in the dataset for several variables. After a thorough review, it has been determined that these outliers represent valid and realistic data points rather than errors or anomalies caused by data entry issues. 

For example:
- High values in `fuel_consumption_city` and `fuel_consumption_comb` likely correspond to vehicles with larger engines or heavy-duty use.
- High `fuel_consumption_comb(mpg)` values may reflect highly fuel-efficient vehicles such as hybrids or electric cars.
- The presence of vehicles with 16 cylinders is rare but plausible for specific types of high-performance or luxury vehicles.

Given that these outliers are genuine data points, removing them would result in a loss of valuable information about the dataset's variability. Therefore, the outliers will be retained for the analysis, ensuring that the results remain comprehensive and inclusive of all relevant cases.

# Train | Test Split

```python
X = df.select_dtypes(include = "number").drop("price",axis=1)
y = df.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

# Implement Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
my_dict= {"Actual": y_test , "pred":y_pred, "residual":y_test - y_pred  }
comparing = pd.DataFrame(my_dict)
comparing
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("---------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```
